Forget about the jargain, these steps can be simplified as:

Step 1- Preprecessing (Extracting SID): 

What it does is to extract the non-sinusoidal part of the current and define it as the disturbance added to the healthy system.

Step 2, 3- Feature Extraction:

What is done here is first we increase the dimentionality of the one dimentional current into two dimentional graphs. This allows us to capture the underlying pattern of the signal.

Then we use some statistical tools, known as joint moments to extract important features of the 2d graphs. Look at joint moments as two dimentional means, standard devition, skewnes,...

The following figure shows current, SID, and two dimentional representation of SID for different disturbances in power system.

```python
img = 'data/fig3.png'
Image(filename=img,width=700, height=500)
```

Step 4- Feature Selection (PCA):

Since we need to run this algorithm on a real-time microprocessor, we need to have a smaller/faster classifier model posible. Therefore, we decreased the size of our features using PCA. The feature size was decreased from 149 to 9. Here is a snapshot of the loss of information by using PVA. With 9 features, we can capture most of information.

```python
img = 'data/PCA.png'
Image(filename=img,width=500, height=500)
```

Step 5- Classifier:

Last step of the pipeline (as of any pipeline), is a classifier.
We compared the result of the following classifiers in terms of accuracy (A) and other metrics that we explain thoroughly in the next notebook.

Comparing all the metrics, we chosen SVM as our classifier.
