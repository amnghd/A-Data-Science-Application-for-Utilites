#### Power System:

"An electric power system is a network of electrical components deployed to supply, transfer, store, and use electric power. "

To put it in understandable words, the process that brings electricity to your outlets is called power systems.


#### Power System Disturbances:

There are so many possible events (disturbances) that would interrupt the transmission and distribution of electricity and would result in large loss for utilities (in order of bilion dollar for hours of interruption).

So how do we detect these disturbances (officially called faults)?
Let's look at the power system as an internet of things. There are many sensors that measure current and voltage (I know!! You might remember the very dull Ohm's law in high school. Or you hopefully might not!!).

For a large portion of the faults, the current and voltage change dramatically and some processors compare their value with a threshold, and BOOM! detect the faults in as fast as 1 ms and disconnect the whole network.

But! There is an important BUT! 

There is a very vermin type of fault that has been "swept under the rug" for decades.

It is known as **High Impedance Fault**. The current and voltage doesn't change that much and also there is a possiblity of fire and explosion. Quite dramatic, right? And also crucial to solve.

Here, take a look at a possible High Impedance Fault that is not being detected (remember faults are usually detected in less than 100 ms).

#### High Impedance Fault:

These type of faults usually occur in distribution system (i.e. possibly closer to residential areas).
In essence, the electricity can not find a pass to close the circut and therefore spikes. However, the spike is not large enough to result in large change in current and voltage.
Therefore, the traditional techniques cannot detect them.

Although, there is not amplitude change, it is shown that the arc results in stochastic changes in the current and voltage.

In a healthy power system current and voltage should be sinusoidal. Due to recent increase in non-linear loads, it is not necessarily sinusoidal anymore. But utilities tr to keep the non-linearity (also known as harmonic distortion) to a low degree.

#### Problem formulation:

So, how can we reformulate this problem into a machine learning issue? We have two categories in power system, anomally (when fault occurs), and healthy (when system is working without anomaly). We also have measurements from current and voltage at a substation (think of it as where the voltage level comes to costumer ready level). 

We can guess what we are looking at: **Classification**!!

We need to see if our current/voltage observation is in Healthy Class, or Anomally class. 

One might think, why don't we use **Anomally detection**. The reason is that one main reason in using anoally detection is that, the anomallies do not necessarily demonstrate a pattern. Therefore, we focus on the normal situation and then whatever else, becomes anomally. However, in power system there are other anomalities (no sinusoial current/voltage) but we don't care enough about them. Therefore, we need to find the pattern in anomallies and compare with the pattern in normal system.

So we are doint a classification on the current and voltae measurement.

Anomally detection in power system is called **power system protection**. In [this article](https://www.sciencedirect.com/science/article/pii/S0378779616304187), we proposed to reformulate all the power system protetion as pattern classification (think of it as a fancy name for data mining when we are concerned about the real time measurements derived from processes). All of the power system protection techniques are in essence Decision Trees that try to discriminate between healthy and faulty situations.
